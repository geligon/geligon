Clean Date Formats

```python
date_columns = ['Artist Begin Date', 'Artist End Date', 'Object End Date','Object Date']

for col in date_columns:
    # Convert to datetime
    df[col] = pd.to_datetime(df[col], errors='coerce')

    # Extract year and replace the original column
    df[col] = df[col].dt.year

    # Convert to integer while handling NaN values
    df[col] = df[col].astype('Int64')

```

```python
df.head()
```

Strip Whitespace

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df.info()
```

```python
# Function to remove blank space entries
def remove_blank_spaces(x):
    if isinstance(x, str):
        return x.strip() if x.strip() else pd.NA
    return x

# Apply the function to all elements in the DataFrame
df = df.map(remove_blank_spaces)

# Remove any rows that are now entirely NA
df = df.dropna(how='all')

# Reset the index if needed
df = df.reset_index(drop=True)
```

```python
object_columns = df.select_dtypes(include=['object']).columns

for column in object_columns:
    df[column] = df[column].fillna('Unknown')
```
